```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo dirigido
G = nx.DiGraph()

# Añadir 8 nodos
G.add_nodes_from(range(8))

# Añadir relaciones dirigidas: cada nodo se conecta a los siguientes dos (ciclo)
for i in range(8):
    G.add_edge(i, (i + 1) % 8) # Conecta al siguiente nodo
    G.add_edge(i, (i + 2) % 8) # Conecta al segundo siguiente nodo

# Visualizar el grafo
plt.figure(figsize=(8, 8))
pos = nx.circular_layout(G) # Usamos un diseño circular para una mejor visualización
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=700)
nx.draw_networkx_edges(G, pos, edge_color='gray', arrowsize=20)
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')
plt.title('Red Dirigida de 8 Nodos')
plt.axis('off') # Ocultar los ejes
plt.show()
```
